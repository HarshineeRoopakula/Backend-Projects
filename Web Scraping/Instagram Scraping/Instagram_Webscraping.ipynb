## Credentials for interacting with Instagram API
For Login into Instagram via instagramy, session id is required. No username or password is needed. You must be login into Instagram via Browser to get session id

- Login into Instagram in default webbrowser
- Move to Inspect option
- Go to Application and then to Storage and then to Cookies and copy the sessionid


```python
credentials_df = pd.read_csv('credentials_instagram.csv',header=None,names=["session_id","id"])
session_id = credentials_df["id"][0]
```

## 1. Extracting details of an Instagram User

```python
from instagramy import InstagramUser
user = InstagramUser("ramit",sessionid=session_id)
user.biography
```

```python
data = user.user_data
```

### Creating a Dataframe for some details 

```python
import pandas as pd
pd.set_option('display.max_colwidth', None)
df = pd.DataFrame(data)
df = df[["biography","full_name","edge_followed_by","edge_follow",'edge_owner_to_timeline_media',"category_name"]]
df = df.rename({"count":""},axis=0)
df = df.drop(["edges","page_info"])
df = df.rename({"edge_followed_by":"number_of_followers","edge_follow":"number_of_followings",'edge_owner_to_timeline_media':"numer_of_posts"},axis=1)
df['number_of_followers'] = df['number_of_followers'].astype(int)
df["number_of_followings"] = df["number_of_followings"].astype(int)
```

```python
df.head()
```
